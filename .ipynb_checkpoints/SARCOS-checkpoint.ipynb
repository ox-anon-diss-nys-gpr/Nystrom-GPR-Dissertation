```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import time
from numpy.linalg import pinv, norm, eigh
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from scipy.spatial.distance import cdist
from scipy.spatial.distance import pdist
import scipy.io as sio
from sklearn.model_selection import KFold
import scipy.linalg as spl
```

```python
# Kernel function for vectors
def gaussian_kernel_vectors(x, y, sigma):
    return np.exp(-np.sum((x - y)**2) / (2 * sigma**2))
```

```python
# Kernel function for matrices
def gaussian_kernel_matrices(X, Y, sigma):
    # Compute squared norms of each row in X and Y
    X_norm_sq = np.sum(X**2, axis=1).reshape(-1, 1)
    Y_norm_sq = np.sum(Y**2, axis=1).reshape(1, -1)
    
    # Compute squared Euclidean distances
    sq_dists = X_norm_sq + Y_norm_sq - 2 * X.dot(Y.T)
    
    # Compute the Gaussian kernel matrix
    K = np.exp(-sq_dists / (2 * sigma**2))
    return K
```

```python
# Uniform sampling
def sample_uniformly(data, num_samples, random_state=None):
    if random_state is not None:
        np.random.seed(random_state)
    indices = np.random.choice(data.shape[0], size=num_samples, replace=False)
    return np.sort(indices)
```

```python
# RPCholesky Sampling: returns (indices, F) so that the Nyström approximation is F @ F.T.
def rpcholesky_kernel(data, k, sigma, random_state=None):
    if random_state is not None:
        np.random.seed(random_state)
    N, d = data.shape
    F = np.zeros((N, k))
    S = np.empty(k, dtype=np.int64)
    
    # Precompute the diagonal entries.
    d_vec = np.array([gaussian_kernel_vectors(x, x, sigma) for x in data])
    
    for i in range(k):
        p = d_vec / d_vec.sum()
        s = np.random.choice(N, p=p)
        S[i] = s
        
        diff = data - data[s]
        g = np.exp(-np.sum(diff**2, axis=1) / (2 * sigma**2))
        
        if i > 0:
            g = g - F[:, :i] @ F[s, :i].T
        
        pivot_val = g[s]
        if pivot_val <= 0:
            raise ValueError(f"Nonpositive pivot value encountered at step {i}.")
        F[:, i] = g / np.sqrt(pivot_val)
        
        F_col_sq = F[:, i]**2
        d_vec = d_vec - F_col_sq
        d_vec = np.maximum(d_vec, 0)
        
    return S, F
```

```python
# K-Means sampling
def kmeans_subset(X, n_clusters, random_state=None):
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    kmeans.fit(X)
    centers = kmeans.cluster_centers_
    indices = []
    # Loop over each center, compute distances from all points, then pick the closest index.
    for center in centers:
        dists = np.linalg.norm(X - center, axis=1)
        indices.append(np.argmin(dists))
    return np.array(indices), centers
```

```python
# Recursive RLS-Nystrom Sampling: returns landmark indices.
def recursiveNystrom(X, n_components: int, sigma, accelerated_flag=False, 
                     random_state=None, lmbda_0=0, return_leverage_score=False, **kwargs):
    rng = np.random.RandomState(random_state)

    n_oversample = np.log(n_components)
    k = np.ceil(n_components / (4 * n_oversample)).astype(int)
    n_levels = np.ceil(np.log(X.shape[0] / n_components) / np.log(2)).astype(int)
    perm = rng.permutation(X.shape[0])

    size_list = [X.shape[0]]
    for l in range(1, n_levels+1):
        size_list.append(np.ceil(size_list[-1] / 2).astype(int))

    sample = np.arange(size_list[-1])
    indices = perm[sample]
    weights = np.ones((indices.shape[0],))

    # Compute the kernel diagonal by evaluating the kernel on each data point with itself.
    k_diag = np.array([gaussian_kernel_vectors(x, x, sigma) for x in X])
    
    for l in reversed(range(n_levels)):
        current_indices = perm[:size_list[l]]
        KS = gaussian_kernel_matrices(X[current_indices, :], X[indices, :], sigma)
        SKS = KS[sample, :]  # sampled rows and columns

        if k >= SKS.shape[0]:
            lmbda = 1e-5
        else:
            lmbda = (np.sum(np.diag(SKS) * (weights ** 2)) -
                     np.sum(spl.eigvalsh(SKS * weights[:, None] * weights[None, :], 
                                          subset_by_index=(SKS.shape[0]-k, SKS.shape[0]-1)))) / k
        lmbda = np.maximum(lmbda_0 * SKS.shape[0], lmbda)
        if lmbda == lmbda_0 * SKS.shape[0]:
            print("Set lambda to %d." % lmbda)

        R = np.linalg.inv(SKS + np.diag(lmbda * weights ** (-2)))
        R = np.matmul(KS, R)

        if l != 0:
            leverage_score = np.minimum(1.0, n_oversample * (1 / lmbda) *
                                        np.maximum(0.0, (k_diag[current_indices] - np.sum(R * KS, axis=1))))
            sample = np.where(rng.uniform(size=size_list[l]) < leverage_score)[0]
            if sample.size == 0:
                leverage_score[:] = n_components / size_list[l]
                sample = rng.choice(size_list[l], size=n_components, replace=False)
            weights = np.sqrt(1.0 / leverage_score[sample])
        else:
            leverage_score = np.minimum(1.0, (1 / lmbda) *
                                        np.maximum(0.0, (k_diag[current_indices] - np.sum(R * KS, axis=1))))
            p = leverage_score / leverage_score.sum()
            sample = rng.choice(X.shape[0], size=n_components, replace=False, p=p)
        indices = perm[sample]

    if return_leverage_score:
        return indices, leverage_score[np.argsort(perm)]
    else:
        return indices
```

```python
# Calculate RMSE
def compute_rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true.ravel() - y_pred.ravel())**2))
```

```python
# Calculate SMSE
def compute_smse(y_true, y_pred):
    mse = np.mean((y_true - y_pred)**2)
    var_y = np.var(y_true)
    return mse / var_y
```

```python
# Calculate SMAPE
def compute_smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred)))
```

```python
def nystrom_approximation_regular(X_train, bandwidth, indices):
    X_m = X_train[indices]
    K_mm = gaussian_kernel_matrices(X_m, X_m, bandwidth)
    K_mn = gaussian_kernel_matrices(X_m, X_train, bandwidth)
    return K_mm, K_mn, X_m
```

```python
def gpr_nystrom(X_train, X_test, y_train, ridge, bandwidth, K_mm, K_mn, X_m, rank):
    A = ridge**2 * K_mm + K_mn @ K_mn.T + 1e-6 * np.eye(rank)
    inverse = np.linalg.inv(A)
    
    K_star_m = gaussian_kernel_matrices(X_test, X_m, bandwidth)
    
    # Compute v = K_mn @ y_train. K_mn is (1000, N_train) and y_train is (N_train,1)
    v = K_mn @ y_train  # shape: (1000,1)
    
    # Compute w = inverse @ v. inverse is (1000,1000) so w is (1000,1)
    w = inverse @ v
    
    # Finally, compute mean = K_star_m @ w. K_star_m is (N_test,1000)
    mean = K_star_m @ w  # shape: (N_test,1)
    return mean
```

```python
# Nystrom approximation regression
def nystrom_regression(X_train, X_test, y_train, bandwidth, ridge, indices):
    X_m = X_train[indices]
    K_mm = gaussian_kernel_matrices(X_m, X_m, bandwidth)
    K_mn = gaussian_kernel_matrices(X_m, X_train, bandwidth)
    A = ridge**2 * K_mm + K_mn @ K_mn.T
    inverse = np.linalg.pinv(A)
    
    K_star_m = gaussian_kernel_matrices(X_test, X_m, bandwidth)
    
    # Compute v = K_mn @ y_train. K_mn is (1000, N_train) and y_train is (N_train,1)
    v = K_mn @ y_train  # shape: (1000,1)
    
    # Compute w = inverse @ v. inverse is (1000,1000) so w is (1000,1)
    w = inverse @ v
    
    # Finally, compute mean = K_star_m @ w. K_star_m is (N_test,1000)
    mean = K_star_m @ w  # shape: (N_test,1)
    return mean
```

```python
# Function for K_mm
def compute_Kmm_inv_sqrt(K_mm):
    # Compute the Cholesky factorization: K_mm = L L^T
    L = np.linalg.cholesky(K_mm)
    # Compute the inverse of L (efficient since L is lower-triangular)
    L_inv = np.linalg.inv(L)
    # Then K_mm^{-1/2} is the transpose of L_inv
    Kmm_inv_sqrt = L_inv.T
    return Kmm_inv_sqrt
```

```python
def nystrom_rp(X_train, X_test, y_train, bandwidth, ridge, indices, F):
    # Extract landmark points
    X_m = X_train[indices]
    
    # Compute K_mm and its inverse square root
    K_mm = gaussian_kernel_matrices(X_m, X_m, bandwidth)
    Kmm_inv_sqrt = compute_Kmm_inv_sqrt(K_mm)  # uses Cholesky
    
    # Instead of computing (F F^T + ridge^2 I_n)^{-1} (which is huge), we work in the low-dimensional space.
    # Form the small matrix: A_small = ridge^2 * I_m + F^T F (m x m)
    m_val = F.shape[1]
    A_small = ridge**2 * np.eye(m_val) + F.T @ F
    A_small_inv = np.linalg.inv(A_small)
    
    # Solve for α using the Woodbury identity:
    # α = 1/ridge^2 * ( y - F @ (A_small_inv @ (F.T @ y)) )
    y_train_vec = y_train.reshape(-1, 1)
    alpha = (1 / ridge**2) * (y_train_vec - F @ (A_small_inv @ (F.T @ y_train_vec)))
    
    # Compute the transformed test kernel vector:
    # F_star = K(X_test, X_m) @ K_mm^{-1/2} (n_test x m)
    K_star_m = gaussian_kernel_matrices(X_test, X_m, bandwidth)
    F_star = K_star_m @ Kmm_inv_sqrt
    
    # Finally, compute the predictive mean: mu = F_star @ (F.T @ α)
    pred_mean = F_star @ (F.T @ alpha)
    
    return pred_mean.ravel()
```

```python
# Preprocessing data
data_dict = sio.loadmat(r"C:\Users\mamun\Downloads\sarcos_inv.mat")
# Assuming the variable is named 'sarcos_inv'
data = data_dict['sarcos_inv']

# Split features and target.
# Here, we assume the last column is the target.
X_train = data[:, :-1]
y_train = data[:, -1]

# Do the same for the test data (if it has the same structure)
test_data_dict = sio.loadmat(r"C:\Users\mamun\Downloads\sarcos_inv_test.mat")
test_data = test_data_dict['sarcos_inv_test']
X_test = test_data[:, :-1]
y_test = test_data[:, -1]

# Standardize the features (fit on train, transform both train and test)
scaler_X = StandardScaler()
X_train_std = scaler_X.fit_transform(X_train)
X_test_std  = scaler_X.transform(X_test)

# Standardize the targets (if doing GP regression, it's common to standardize them as well)
scaler_y = StandardScaler()
y_train_std = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
y_test_std  = scaler_y.transform(y_test.reshape(-1, 1)).ravel()
```

```python
# Hyperparameter candidates
ridge_list = [1e-2, 1e-1]
bandwidth_list = [49, 50, 51, 52] 
ranks = [16, 32, 64, 128, 256, 512, 1024]
cv_folds = 3

# Subsample data for fast CV (optional but recommended for YearPredictionMSD)
subset_size = 10000
X_sub = X_train_std[:subset_size]
y_sub = y_train_std[:subset_size]

results = []

for rank in ranks:
    print(f"\n=== Rank: {rank} ===")
    
    for ridge in ridge_list:
        for bandwidth in bandwidth_list:
            rmse_scores = []

            kf = KFold(n_splits=cv_folds, shuffle=True, random_state=42)
            for train_index, val_index in kf.split(X_sub):
                X_tr, X_val = X_sub[train_index], X_sub[val_index]
                y_tr, y_val = y_sub[train_index], y_sub[val_index]

                indices = sample_uniformly(X_tr, rank)
                K_mm, K_mn, X_m = nystrom_approximation_regular(X_tr, bandwidth, indices)
                y_pred = gpr_nystrom(X_tr, X_val, y_tr, ridge, bandwidth, K_mm, K_mn, X_m, rank)

                rmse = compute_rmse(y_val, y_pred)
                rmse_scores.append(rmse)

            avg_rmse = np.mean(rmse_scores)
            print(f"Ridge={ridge}, Bandwidth={bandwidth} → Avg RMSE={avg_rmse:.4f}")
            results.append((rank, ridge, bandwidth, avg_rmse))

# Rank all combinations
results.sort(key=lambda x: x[3])  # sort by RMSE
print("\n=== Best Configurations (Ranked by RMSE) ===")
for rank, ridge, bandwidth, rmse in results:
    print(f"Rank={rank}, Ridge={ridge}, Bandwidth={bandwidth} → RMSE={rmse:.4f}")
```

```python
# Nyström Approximation: given the full kernel matrix and landmark indices.
def nystrom_approximation(K, sampled_indices):
    K_sample = K[:, sampled_indices]
    K_SS = K[np.ix_(sampled_indices, sampled_indices)]
    K_SS_pinv = np.linalg.pinv(K_SS)
    return K_sample @ K_SS_pinv @ K_sample.T
```

```python
# Parameters
sigma = 49
ranks = [16, 32, 64, 128, 256, 512, 1024]
num_trials = 10
methods = ['Uniform', 'Recursive', 'RPCholesky', 'K-Means++']

errors = {method: [] for method in methods}

for r in ranks:
    print(f"Processing Rank {r}...")
    err_uniform, err_recursive, err_rpchol, err_kmeans = [], [], [], []

    for t in range(num_trials):
        print(f"Trial {t+1} / {num_trials}")

        # Randomly sample 10,000 points
        idx_sample = np.random.choice(X_train_std.shape[0], 10000, replace=False)
        X_sample = X_train_std[idx_sample]

        # Compute full kernel using your function
        K_full = gaussian_kernel_matrices(X_sample, X_sample, sigma)
        full_trace = np.trace(K_full)

        # Uniform Sampling
        idx = sample_uniformly(X_sample, r, random_state=t)
        K_approx = nystrom_approximation(K_full, idx)
        err_uniform.append(np.abs(full_trace - np.trace(K_approx)) / full_trace)

        # Recursive RLS
        idx = recursiveNystrom(X_sample, n_components=r, sigma=sigma, random_state=t)
        K_approx = nystrom_approximation(K_full, idx)
        err_recursive.append(np.abs(full_trace - np.trace(K_approx)) / full_trace)

        # RPCholesky
        idx, F = rpcholesky_kernel(X_sample, r, sigma, random_state=t)
        K_approx = F @ F.T
        err_rpchol.append(np.abs(full_trace - np.trace(K_approx)) / full_trace)

        # KMeans++
        idx, _ = kmeans_subset(X_sample, n_clusters=r, random_state=t)
        K_approx = nystrom_approximation(K_full, idx)
        err_kmeans.append(np.abs(full_trace - np.trace(K_approx)) / full_trace)

    # Store mean error over trials
    errors['Uniform'].append(np.mean(err_uniform))
    errors['Recursive'].append(np.mean(err_recursive))
    errors['RPCholesky'].append(np.mean(err_rpchol))
    errors['K-Means++'].append(np.mean(err_kmeans))

# Plotting
plt.figure(figsize=(8, 6))
for method in methods:
    plt.plot(ranks, errors[method], marker='o', label=method)

plt.xlabel('Rank')
plt.ylabel('Average Relative Trace Error')
plt.yscale('log')
plt.title('Relative Trace Error vs Rank (YearPredictionMSD, Bandwidth=8.0)')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```
